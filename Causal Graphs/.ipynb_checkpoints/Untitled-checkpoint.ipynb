## Importing Library

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from IPython.display import display, Image
```

## Creating data

```python
A = np.linspace(0,50,100)
B = np.linspace(100,500,100)
C= A*1 + B*5
```

## DAG for the data

```python
display(Image(filename='ABC_model.jpg'))
```

```python
df = pd.DataFrame({"A" : A, "B" : B, "C" : C})
df
```

## OLS regression Result

```python
ols = smf.ols("C ~ A+B", df).fit()
ols.summary()
```
